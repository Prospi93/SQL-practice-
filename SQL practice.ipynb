```python
squadre_e_stadio = %sql select team, stadium from CALCIO_MONDO group by stadium;
squadre_e_stadio
```

```python
goals_wins = %sql select season_goals as 'season_goals', count(*) as 'frequency' from CALCIO_MONDO group by season_goals;
goals_wins
```

```python
df = goals_wins.DataFrame() 
%matplotlib inline 
import seaborn 
plot=seaborn.barplot(x='season_goals', y='frequency', data=df)
```

```python
nazione_distribuzione = %sql select nazione as 'nazione', count(*) as 'frequency' from CALCIO_MONDO group by nazione;
nazione_distribuzione
```

```python
df2 = nazione_distribuzione.DataFrame()
```

```python
%matplotlib inline 
plot2 = seaborn.barplot(x='nazione', y='frequency', data=df2)
```
